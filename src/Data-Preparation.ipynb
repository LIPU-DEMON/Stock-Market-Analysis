```python
import pandas as pd
from zipfile import ZipFile
import yfinance as fp
import datetime
```

```python
today = datetime.date.today()
year  = today-datetime.timedelta(days=7300) #20 Years
Stocks = ['AAPL', 'MSFT', 'IBM', 'GE', 'JNJ']
data = fp.download(Stocks,start=year,end=today,group_by="ticker")
data
```

```python
new = []
for i in Stocks:
    df = data[i].copy()
    df = df.reset_index()
    df["Stocks"] = i
    new.append(df)
new_data = pd.concat(new)
new_data  
```

```python
new_data.info()
```

```python
#Saving File
new_data.to_csv("Data/DataPreparation.csv",index=False)
```
